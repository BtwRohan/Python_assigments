```python
print('Welcome to Library, what you want to do ?')
print('1. Add Book')
print('2. Remove Book')
print('3. Search Book')
print('4. List All Book')
print('5. Catogarize Book')
print('6. Check Duplicates')
print('7. Exit')

BookList = []

f = open('LibraryBooks.txt','r+')
for i in f:
    data = i.split(',')
    TupleData = tuple(data)
    BookList.append(TupleData)

def add_book(title,author,genre):
    return (title,author,genre)
    
def FileSaving(BookList,f):
    f.seek(0)
    f.truncate(0)
    for title,author,genre in BookList:
        f.write(f'{title},{author},{genre}\n')

def add_to_library(book):
    BookList.append(book)
    FileSaving(BookList,f)
    print(book[0],'added successfully')

def remove_from_book():
    DeleteBook = input('Enter the title of book to remove: ')
    found = False
    
    for i in BookList:
        
        if DeleteBook == i[0]:
            BookList.remove(i)
            found = True
            FileSaving(BookList,f)
            break

    if not found:
        print('\nBook does not exist')

def search_book():
    SearchBook = input('Enter title or author name of the Book: ')
    found = False
    print('Book Matching')
    
    for i in BookList:
        if SearchBook.lower() in i[0].lower or SearchBook.lower() in i[1].lower:
            print(f'Title: {i[0]}, Author: {i[1]}, Genre: {i[2]}')
            found = True

    if not found:
        print('\nNo Match Found')
            


def list_book():
    for index,book in enumerate(BookList,start=1):
        title,author,genre = book
        print(f'{index}, Title: {title}, Author: {author}, Genre: {genre}')
    print()

def catogarize_book():
    BookCatogary = {}
    
    for title,author,genre in BookList:
    
        if genre not in BookCatogary:
            BookCatogary[genre]=[]
    
        BookCatogary[genre].append((title, author))
    
    print('Catogarized Books: ')
    
    for genre, i in BookCatogary.items():
        print(f"\nGenre: {genre}")
        
        for book in i:
            print(f"Title: {book[0]}, Author: {book[1]}")
    print()

def check_duplicate(title):
    
    for book in BookList:
        
        if title.lower() == book[0].lower:
            return True
    
    return False    

def ListClear():
    BookList.clear()
    f.seek(0)
    f.truncate(0)

while True:
    ch = int(input(f'Enter choice: ')) 
    
    if ch == 1:
        book = add_book(input('Enter book title: '),input('Enter Author: '),input('Enter genre: '))
        choice = input('Add Book in library (Y/N): ').lower()
    
        if choice == 'y':
    
            if check_duplicate(book[0]):
                print('\n\nBook already exist in list')
    
            else:
                add_to_library(book)
    
        else:
            print()
    
    elif ch == 2:
        list_book()
        remove_from_book()
        print('\nAfter removing the book:\n')
        list_book()
    
    elif ch == 3:
        search_book()
    
    elif ch == 4:
        print('Book List: ')
        list_book()
    
    elif ch == 5:
        catogarize_book()
    
    elif ch == 6:
        true = check_duplicate(input(f'Enter the book you want to check for duplicate copy: '))
        if true:
            print(book[0],' Exists')
        else:
            print(book[0],'is addable')

    elif ch == 7:
        choice2 = input('Clear List Surely (Y/N): ').lower()
        if choice2 == 'y':
            ListClear()
        else:
            continue

    elif ch == 8:
        print(f'\nLibrary Exited')
        break
    
    else:
        print(f'\nInvalid choice')
```
